# calcualte_angle and defs

```python
def calculate_angle(a,b,c):
    a=np.array(a)
    b=np.array(b)
    c=np.array(c)
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1],a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle>180.0:
        angle = 360-angle
    return angle

def calculate_y(a,b):
    a=np.array(a)
    b=np.array(b)
    
    dis = abs(a[1]-b[1]) *100
    return dis

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedir(directory)
    except OSError:
        print('Error: Creating directory. ' + directory)
        
```

```python
def extract_keypoints(results):
    pose =[]
    for res in results.pose_landmarks.landmark:
        test = np.array([res.x, res.y,res.z])
        if(res.visibility > 0.9):
            pose.append(test)
            
    return pose
```

```python
import cv2
import mediapipe as mp
import numpy as np
import pygame
import time
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
pygame.mixer.init()
cap =cv2.VideoCapture(0)
##setup mediapipe instance
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    for action in actions:
        mySound = pygame.mixer.Sound(action+".mp3")
        mySound.play()
        #LOOP through sequneces aka videos
        for sequence in range(no_sequences):
            #Loop through video length aka sequnece length
            for frame_num in range(sequence_length):
                
                #Read feed
                ret,frame = cap.read()

                #Recoloring TI RGB
                image = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
                image.flags.writeable = False

                #make detection
                results = pose.process(image)

                #recoloring back to BGR
                image.flags.writeable = True
                image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
                
                # Render detections
                mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
                
                if frame_num == 0:
                    cv2.putText(image, 'STARTING COLLECTION', (120,200),
                                cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 4, cv2.LINE_AA)

                    cv2.putText(image, 'Collectiong frames for {} Video Number {}'.format(action,sequence), (15,12),
                                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 4, cv2.LINE_AA)
                    #show to screen
                    cv2.imshow('OpenCV Feed',image)
                    cv2.waitKey(2000)
                else:
                    cv2.putText(image, 'Collectiong frames for {} Video Number {}'.format(action,sequence), (15,12),
                                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 4,cv2.LINE_AA)

                    #show to screen
                    cv2.imshow('OpenCV Feed',image)
                
                #keypoints input
                keypoints = extract_keypoints(results)
                img_path = os.path.join(DATA_PATH,action,str(sequence),str(frame_num))
                cv2.imwrite(img_path+'.jpg',frame, params=[cv2.IMWRITE_PNG_COMPRESSION,0])
                npy_path=os.path.join(DATA_PATH, action, str(sequence),str(frame_num))
                np.save(npy_path,keypoints)    
                
                #breaking
                if cv2.waitKey(10) & 0xFF == ord('q'):
                    break
        
cap.release()
cv2.destroyAllWindows()
```

```python
import os
import cv2
import mediapipe as mp
import numpy as np
import time
import pygame

#pose 분석 초기화
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

#카메라 초기화
cap =cv2.VideoCapture(0)

#음성출력 초기화
pygame.mixer.init()

#counter
counter_front=0
counter_side1=0
counter_side2=0
counter_dis=0
counter_good=0

stage_front = None
stage_side = None
stage_side = None
stage_shoulder = None

#make file directory
createFolder('good_pose')
createFolder('bad_front')
createFolder('bad_side1')
createFolder('bad_side2')
createFolder('bad_shoulder')

##setup mediapipe instance
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret,frame = cap.read()
        
        #Recoloring TI RGB
        image = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        
        #make detection
        results = pose.process(image)
        
        #recoloring back to BGR
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        #Extract landmakrs
        try:
            landmarks = results.pose_landmarks.landmark
            #Get coordinates 시작, 중간, 끝으로 각도를 계산한다.
            shoulder_L = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x ,landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
            shoulder_R = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x ,landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
            nose = [landmarks[mp_pose.PoseLandmark.NOSE.value].x ,landmarks[mp_pose.PoseLandmark.NOSE.value].y]
            wirst = [landmarks[mp_pose.PoseLandmark.NOSE.value].x ,landmarks[mp_pose.PoseLandmark.NOSE.value].y]
            
            #calculate angle
            angle = calculate_angle(shoulder_L , shoulder_R , nose)
            angle2 = calculate_angle(shoulder_R , shoulder_L , nose)
            
            #왼쪽 어깨와 오른쪽 어깨 수치의 차이
            dis = calculate_y(shoulder_L , shoulder_R)
            
            #visualize
            cv2.putText(image, str(angle),
                        tuple(np.multiply(shoulder_R,[640,480]).astype(int)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255), 2, cv2.LINE_AA
                             )     
            
            cv2.putText(image, str(angle2),
                        tuple(np.multiply(shoulder_L,[640,480]).astype(int)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255), 2, cv2.LINE_AA
                             )     
            
            cv2.putText(image, str(dis),
                        tuple(np.multiply(nose,[640,480]).astype(int)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255), 2, cv2.LINE_AA
                             )     
            
            #counter logic
            #bad_front angle
            if angle <50 and counter_front<200:
                cv2.imwrite('bad_front/%s%d.png' %("bad_front",counter_front),frame, params=[cv2.IMWRITE_PNG_COMPRESSION,0])
                counter_front += 1
            #bad_side1 angle        
            elif angle>70 and counter_side < 200:
                cv2.imwrite('bad_side/%s%d.png' %("bad_side",counter_dis),frame, params=[cv2.IMWRITE_PNG_COMPRESSION,0])
                counter_side+=1    
            #bad_side1 angle 
            elif angle2>70 and counter_side2 < 200:
                cv2.imwrite('bad_side2/%s%d.png' %("badside_2",counter_dis),frame, params=[cv2.IMWRITE_PNG_COMPRESSION,0])
                counter_side2+=1
                
            #bad shoulder
            elif dis >2 and counter_dis < 200:
                cv2.imwrite('bad_shoulder/%s%d.png' %("bad_dis",counter_dis),frame, params=[cv2.IMWRITE_PNG_COMPRESSION,0])
                counter_dis+=1
                
            #good_pose    
            elif angle>52 and counter_good<200:
                cv2.imwrite('good_pose/%s%d.png' %("good_pose",counter_good),frame, params=[cv2.IMWRITE_PNG_COMPRESSION,0])
                counter_good+=1
            
        except:
            pass
        
        #setup status box
        cv2.rectangle(image,(0,0,),(225,73),(245,117,16),-1)
        
        #rep data
        cv2.putText(image,'REPS',(15,12),
                   cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,0,0),1,cv2.LINE_AA)
        
        cv2.putText(image,str(counter_front),(10,60),
                   cv2.FONT_HERSHEY_SIMPLEX,2, (255,255,255),2,cv2.LINE_AA)
        
        # Render detections
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
        
        cv2.imshow('Mediapipe Feed2',image)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
